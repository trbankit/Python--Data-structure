#Data types and structures questions

1. What are data structures, and why are they important?
- > Data structures are ways of organizing and storing data to perform operations efficiently. They're important because they help manage and process large amounts of data quickly.

2. Explain the difference between mutable and immutable data types with examples.
- > Mutable data types can be changed after creation (like lists, sets, dictionaries). Immutable data types cannot be changed after creation (like strings, tuples, integers).

3. What are the main differences between lists and tuples in Python?
- > Lists are mutable and can be modified; tuples are immutable. Lists use square brackets, tuples use parentheses.

4. Describe how dictionaries store data.
- > Dictionaries store data in key-value pairs. Each key is unique and maps to a value.

5. Why might you use a set instead of a list in Python?
- > Sets are used to store unique, unordered data. They're faster for membership tests compared to lists.

6. What is a string in Python, and how is it different from a list?
- > A string is a sequence of characters. It's immutable, meaning it cannot be changed after creation.

7. How do tuples ensure data integrity in Python?
- > Tuples are immutable, so their contents can't be altered after creation, ensuring integrity.

8. What is a hash table, and how does it relate to dictionaries in Python?
- > A hash table maps keys to values using a hash function. Dictionaries in Python are implemented using hash tables.

9. Can lists contain different data types in Python?
- > Yes, lists can store different data types like int, str, and float in one list.

10. Explain why strings are immutable in Python.
- > Strings are immutable in Python for security, consistency, and optimization purposes.

11. What advantages do dictionaries offer over lists for certain tasks?
- > Dictionaries offer faster lookup times for keys compared to searching through a list.

12. Describe a scenario where using a tuple would be preferable over a list.
- > Use a tuple when data shouldn't change, like days of the week or constant coordinates.

13. How are keys and mutable values in Python?
- > Dictionaries map unique keys to values, unlike lists which use indices.

14. How does the 'in' keyword work differently for lists and dictionaries?
- > 'in' checks for an element in a list but for a key in a dictionary.

15. How do sets handle the elements of a tuple? Explain why or why not.
- > Tuple elements are set at creation. You cannot add or remove them later.

16. What is a nested dictionary, and give an example of its use case?
- > A nested dictionary has a dictionary inside another dictionary, e.g. {'student': {'name': 'John', 'age': 20}}.

17. Describe the time complexity of accessing elements in a dictionary
- > Time complexity for dictionary lookup is O(1) on average due to hashing.

18. In what situations are sets preferred over dictionaries?
- > Sets are used when only unique items are needed and order doesn't matter.
19. Why are dictionaries considered unordered, and how does that affect data retrieval?
- > Dictionaries are unordered because the order of insertion isn't preserved; they're optimized for quick lookup.

20. Explain the difference between a list and a dictionary in terms of data retrieval.
- > Lists retrieve data by index in O(1) time. Dictionaries retrieve by key, usually in O(1) time too, but unordered.
